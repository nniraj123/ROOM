```python
##################################################
######### Method: SLR + Additive Noise  ##########
##################################################
import numpy as np
import matplotlib.pyplot as plt
import math
import json
from netCDF4 import Dataset
from functionsLR_AddNoise import *
from scipy import fftpack
from datetime import datetime
import multiprocessing
from joblib import Parallel, delayed
from tqdm import tqdm
# np.random.seed(42)

num_cores = multiprocessing.cpu_count()
print('Number of cores:',num_cores)

# Define other parameters of the problem
n_pc = 150         # number of PCs
n_record = 50000   # total records
n_maxtrain = 40000 # maximum possible length of training data
n_test = n_record - n_maxtrain # length of the test data
test0_idx = 40000              # test data start index
dt = 10            # temporal resolution

#### Load the pcs ####
f = '../psi1_DG_0_500K_100days_filtered_150PCs.dat'
pcs = dlmread(f,n_record,n_pc)
pcs = center(pcs)
# get the test dataset
test_data = pcs[test0_idx:test0_idx+n_test, :]
 
#### Load the EOFs ####
nx = 513; ny=513
eoffile = os.getcwd() + '/../psi1_DG_0_500K_100days_filtered_150EOFs.nc'
fid = Dataset(eoffile,'r')
eofs = fid.variables['EOFs'][:n_pc,:,:]
eofs = eofs.reshape((n_pc, nx*ny))
```

```python
### Red and White noise definitions ###

def l2_norm(array):
    return np.sqrt(np.sum(np.power(array,2)))

# Red Noise
def corrRedNoise(n_samples, n_in_param, residual, tau, deltaT):
    
    eta = np.zeros((n_samples, n_in_param))
    
    ## compute std of the white noise
    var_res = np.var(residual, axis=0)
    gamma = 1/tau 
    c = (1 - gamma*deltaT)
    sigma = np.sqrt(var_res*(1 - c**2)/(deltaT**2))
    
    # generate white noise and spinup the red noise system to draw 
    # samples from the correct distribution
    n_spin = 100000
    eta_ = np.zeros((n_spin, n_in_param))
    for i in range(1,n_spin):
        zeta_ = sigma*np.random.randn(1, n_in_param)
        eta_[i,:] = c*eta_[i-1,:] + deltaT*zeta_
        
    # compare the std of eta_ with that of the residuals
    std_diff = np.sqrt(var_res) - np.std(eta_, axis=0)
    print('L2 norm of the error in std(red noise):', l2_norm(std_diff))
    
    eta[0,:] = eta_[-1,:]
    for i in range(1, n_samples): 
        zeta = sigma*np.random.randn(1, n_in_param)
        eta[i,:] = c*eta[i-1,:] + deltaT*zeta
    
    covn = np.corrcoef(residual.T) # Correlation coefficients
    rr = np.linalg.cholesky(covn)
    
    return eta.dot(rr.T)

##################################################################

# Correlated white noise
def corrWhtNoise(nx, ny, residual):
    dW = np.random.randn(nx, ny) # additive white noise
    covn = np.corrcoef(residual.T) # Correlation coefficients
    rr = np.linalg.cholesky(covn)
    stdres = np.std(residual, axis=0)
    return dW.dot(rr.T)*stdres
```

```python
#################################################################
######### Calculation of decorrelation time for each PC #########
#################################################################
num_lags = 1200
tau = np.zeros((n_pc,))

start = datetime.now()

for k in range(n_pc):
    autocor_x = tsaplots.acf(pcs[:,k], nlags=num_lags, fft=False)
    tau[k] = np.where(autocor_x<1/np.exp(1))[0][0]

tau = tau*dt
print('Tau calculation time = ',datetime.now()-start)
print(tau)

#################################################################
### Sensitivity analysis of training length to the performance ##
#################################################################
# check the RMSE on the test data for next time step forecast

n_mintrain = 1000
n_train_inc = 1000

n_train = np.arange(n_mintrain, n_maxtrain+1, n_train_inc, dtype='int')
rmse_test_whtnoise = np.zeros((len(n_train), n_pc)) 
rmse_test_rednoise = np.zeros((len(n_train), n_pc)) 

for i, trainlen in enumerate(n_train):
    print('Training length:',trainlen)
    # prepare the training data
    x_train, y_train, scaler = training4tendency(pcs, n_pc, trainlen, dt, 'standard_top_pc')
    # perform LR
    [linregmodel, reg_res,_] = linear_regression(x_train, y_train)
    # check the performance on the test dataset
    normalized_test_data = scaler.transform(test_data)
    ytendLR_test =  linregmodel.predict(normalized_test_data)
    #### white noise ####
    dW = corrWhtNoise(n_test, n_pc, reg_res)
    yLR_test = ((ytendLR_test+dW)*dt) + normalized_test_data
    # compute the rms error
    rmse_test_whtnoise[i,:] = np.sum((yLR_test[:-1,:] - normalized_test_data[1:,:])**2, axis=0) 
    #### red noise ####
    dW = corrRedNoise(n_test, n_pc, reg_res, tau, dt)
    yLR_test = ((ytendLR_test+dW)*dt) + normalized_test_data
    # compute the rms error
    rmse_test_rednoise[i,:] = np.sum((yLR_test[:-1,:] - normalized_test_data[1:,:])**2, axis=0) 

# Analyze RMSE on the test dataset for each training length
# white noise
#for i in range(n_pc):
#    plt.scatter(n_train, rmse_test_whtnoise[:,i], color='b', alpha=0.1)
plt.plot(n_train, np.mean(rmse_test_whtnoise, axis=1),'r--', linewidth=2)
plt.xlabel('Training length')
plt.ylabel('RMSE')
#plt.title('Sensitivity analysis for the White noise')
plt.xlim([0, 40000])
plt.grid()
plt.show()
# red noise
#for i in range(n_pc):
#    plt.scatter(n_train, rmse_test_rednoise[:,i], color='b', alpha=0.1)
plt.plot(n_train, np.mean(rmse_test_rednoise, axis=1),'r--', linewidth=2)
plt.xlabel('Training length')
plt.ylabel('RMSE')
#plt.title('Sensitivity analysis for the Red noise')
plt.xlim([0, 40000])
plt.grid()
plt.savefig('Mean_sensitivity_to_training_length_LR_for_150PCs.png')
plt.show()
```

```python
# train the LR
n_train = n_maxtrain # Length used for training
[x_train, y_train, scaler] = training4tendency(pcs, n_pc, n_train, dt, 'standard_top_pc')

# Train Linear Regression
start = datetime.now()
[linregmodel, reg_res, r2_sco] = linear_regression(x_train,y_train)
std_res = np.std(reg_res, axis=0)
print('Linear Regression training time:',datetime.now()-start)
```

```python
# plot LR output vs the truth
y_tend_LR = linregmodel.predict(x_train)
for k in range(10):
    plt.plot(y_train[:2000,k])
    plt.plot(y_tend_LR[:2000,k])
    plt.show()
```

```python
#################################################################
######### Calculation of decorrelation time for each PC #########
#################################################################
num_lags = 1200
tau = np.zeros((n_pc,))

start = datetime.now()

for k in range(n_pc):
    autocor_x = tsaplots.acf(x_train[:,k], nlags=num_lags, fft=False)
    tau[k] = np.where(autocor_x<1/np.exp(1))[0][0]

tau = tau*dt
print('Tau calculation time = ',datetime.now()-start)
print(tau)
```

```python
####################################################################
##################### Short term forecasts #########################
####################################################################
n_maxlead = 10  # maximum lead time for the forecasts
n_ic = int((n_test - n_maxlead))    # Number of initial conditions 
n_ensem = 100   # Number of stochastic ensembles for each I.C.
y_pred = np.zeros([n_ic, n_ensem, n_maxlead, n_pc])
noisetype = 'white'

start = datetime.now()

# generate noise
if noisetype=='white':
    dW = corrWhtNoise(n_ic*n_maxlead*n_ensem, n_pc, reg_res)
elif noisetype=='red':
    dW = corrRedNoise(n_ic*n_maxlead*n_ensem, n_pc, reg_res, tau, dt)
else:
    raise ValueError('noise type not supported')

for i in range(n_ic):
    # initial condition
    x_start = test_data[i,:]

    y_pred[i,:,:,:] = forecast(n_ensem, n_maxlead, n_pc, x_start,
                               dt, linregmodel, scaler, 
                               dW[i*n_maxlead*n_ensem:(i+1)*n_maxlead*n_ensem])
    
# Ensemble mean for each IC
enMean_y_pred = np.mean(y_pred, axis=1)

print('Prediction time:', datetime.now()-start)
print('Number of initial conditions:', n_ic)
print('Number of ensembles per initial condition:', n_ensem)
print('Length of each forecast:', n_maxlead)
```

```python
# Plot the predicted results against the truth for any initial condition
ic = 0       # IC index; should be 0,1,2,3..., nic-1 
for k in range(20):
    plt.figure()
    x = np.arange(0,n_maxlead*dt,10)
    plt.plot(x,enMean_y_pred[ic,:,k],label='Additive noise')
    true_ = test_data[ic+1:ic+n_maxlead+1,k]
    plt.plot(x,true_,label='Truth')
    plt.xlabel('Time (in days)')
    plt.title('PC#'+str(k+1))
    plt.legend()
    plt.show()
```

```python
####################################################################
############ Metric 1 : RMSE on the physical space #################
###### Use each data point of the test dataset as an IC ############
## Obtain 100 stochastic realizations (if applicable) for each IC ##
####################################################################
# Project the modelled PCs onto the EOFs and calculate mean RMSE
rmse = np.zeros((n_ic, n_maxlead))

start = datetime.now()
# start a parallel pool and implement thread parallelism
if __name__ == "__main__":
    rmse = Parallel(n_jobs=num_cores, prefer="threads")(delayed(RMSE)(i,test_data,enMean_y_pred,eofs,n_maxlead) for i in tqdm(range(n_ic)))

meanRMSE = np.mean(rmse, axis=0)
```

```python
# Plot the mean RMSE
plt.plot(np.arange(dt,n_maxlead*dt+1,dt),meanRMSE)
plt.xlabel('Time (in days)')
plt.ylabel('RMSE')
plt.ylim([10, 120])
plt.xlim([dt, n_maxlead*dt])
plt.grid(color='k', linestyle='--', linewidth=0.2)
plt.savefig('RMSE_Psi1_50days_predictions_AddRedNoise_npc=150_nic=9995_nensem=100_ntrain=400K_ntest=100K.png',dpi=100)
plt.show()
np.save('RMSE_Psi1_50days_predictions_AddRedNoise_npc=150_nic=9995_nensem=100_ntrain=400K_ntest=100K',meanRMSE)
```

```python
################################################################
#################### Long term forecasts #######################
################################################################
n_maxlead = 20000  # maximum lead time for the forecasts
n_ic = 1    # Number of initial conditions 
n_ensem = 1   # Number of stochastic ensembles for each I.C.
y_pred = np.zeros([n_ic, n_ensem, n_maxlead, n_pc])
noisetype = 'red'

# generate noise
start = datetime.now()
if noisetype=='white':
    dW = corrWhtNoise(n_maxlead*n_ensem, n_pc, reg_res)
elif noisetype=='red':
    dW = corrRedNoise(n_maxlead*n_ensem, n_pc, reg_res, tau, dt)
else:
    raise ValueError('noise type not supported')

print('Noise generation time:', datetime.now()-start)

start = datetime.now()
for k in range(n_ic):
    #### Start Forecasts ####
    x_start = test_data[k,:]
    
    # Forecast
    y_pred[k,:,:,:] = forecast(n_ensem,n_maxlead,n_pc,x_start,
                               dt,linregmodel,scaler,dW)

# Ensemble mean for each IC
enMean_y_pred = np.mean(y_pred, axis=1)

print('Prediction time:', datetime.now()-start)
print('Number of initial conditions:',n_ic)
print('Number of ensembles per initial condition:',n_ensem)
print('Length of each forecast:',n_maxlead)
```

```python
###### Time complexity ######
# For White Noise:
# Training time = LR training time(0.5sec) + time to calculate the standard deviations of LR residuals (0.2sec)
# Prediction time = time to produce one 10K days realization using one initial condition(0.2sec) + noise generation time (0.1)

# For Red Noise:
# Training time = LR training time(0.5sec) + time to compute the red noise parameters (mainly Tau: ~60sec)
# Prediction time = noise spin-up time (3sec) + time to produce one 10K days realization using one initial condition (0.2sec)
```

```python
################################################################
############# Metric 3 : Climatology and Variance ##############
######### using multiple realizations of 10,000 days  #########
################################################################
psi1_climatology_true = np.zeros((n_ic,nx*ny))
psi1_variance_true = np.zeros((n_ic,nx*ny))
psi1_climatology_pred = np.zeros((n_ic,nx*ny))
psi1_variance_pred = np.zeros((n_ic,nx*ny))

# Project the modelled PCs onto the EOFs and calculate climatology
for kk in range(n_ic):
    psi1_true = test_data[kk+1:kk+n_maxlead+1,:].dot(eofs)
    psi1_climatology_true[kk,:] = np.mean(psi1_true, axis=0)
    psi1_variance_true[kk,:] = np.var(psi1_true, axis=0)
    
    psi1_pred = enMean_y_pred[kk,:,:].dot(eofs)
    psi1_climatology_pred[kk,:] = np.mean(psi1_pred, axis=0)
    psi1_variance_pred[kk,:] = np.var(psi1_pred, axis=0)

mean_psi1_climatology_pred = np.mean(psi1_climatology_pred, axis=0)
mean_psi1_variance_pred = np.mean(psi1_variance_pred, axis=0)
mean_psi1_climatology_true = np.mean(psi1_climatology_true, axis=0)
mean_psi1_variance_true = np.mean(psi1_variance_true, axis=0)
```

```python
############################################################
####### Analysis of the tendency and state residuals  ######
############################################################
# Integrate the linear regression output and compute residual
y_pred_LR = (y_tend_LR*dt) + x_train
y_pred_res = scaler.transform(pcs)[1:n_train,:] - y_pred_LR

for k in range(2):
    plt.plot(y_pred_res[:1000, k], label='State Residual')
    plt.plot(reg_res[:1000, k], label='Tendency Residual')
    plt.title('State and tendency residuals (PC#'+str(k+1)+')')
    plt.legend()
    plt.show()

###### Diagnose autocorrelations of both the residuals
n_maxlag = 1000
acorr_tendres = tsaplots.acf(reg_res[:,-1], nlags=n_maxlag, fft=False)
acorr_stateres = tsaplots.acf(y_pred_res[:,-1], nlags=n_maxlag, fft=False)

plt.plot(acorr_tendres[:50], label='Tendency')
plt.plot(acorr_stateres[:50], label='State')
plt.legend()
plt.show()

##### Diagnose frequency contents in both the residuals
fsample = 1/10 # sampling frequency in 1/day

#Perform FFT
fft_tendres = fftpack.fft(reg_res[:,-1])
fft_stateres = fftpack.fft(y_pred_res[:,-1])

#Compute the frequencies
length = len(reg_res)
freq = fftpack.fftfreq(length)*fsample

specPow_tendres = abs(fft_tendres[:int(length/2)])**2
plt.plot(freq[:int(length/2)], specPow_tendres, label='Tendency Res')
#plt.plot(freq[:int(length/2)], fft_stateres[:int(length/2)], label='State Res')
plt.legend()
plt.show()
```
